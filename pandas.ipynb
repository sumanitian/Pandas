# Grouping

```python
df.groupby('year')[['life_exp','gdp_cap']].mean()
```

```python
df.groupby('year')[['life_exp','gdp_cap']].count()
```

```python
df.groupby(['continent','year'])[['life_exp','gdp_cap']].mean()
```

```python
df.groupby(['year','continent'])[['life_exp','gdp_cap']].mean()
```

```python
dfgb = df.groupby(['continent','year'])[['life_exp','gdp_cap']].mean()
dfgb.reset_index()
```
